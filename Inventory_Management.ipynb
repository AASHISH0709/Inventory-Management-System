# For Customer

```python
import json
fd = open("record.json",'r')
txt = fd.read()
fd.close()
recods = json.loads(txt)

fdd= open("sales.json",'r')
txtt = fdd.read()
stock = json.loads(txtt)
fdd.close()
```

```python
print("Welcome To Our Shop")
print("List of product we have :-")
##Customer Name and Moblile Input 
u_name=input("Please Enter Your Name")
U_mob=input("Please Enter Mobile Num")

u_list=[] ##list To store use items
for i in recods: 
    print("Product -" ,recods[i]['Name'],",","Product id :",i,"," , "Unit Price - ₹",recods[i]["Unit_Price"],"/",recods[i]["Unit"] )
print("Now Please Order Items")
amount=0
while(True):
    while(True):
        Ui_code=input("Please Enter Id of Prodect to order item")
        check1=Ui_code in recods.keys()
        if(check1==True):
            break
        else:
            print("Product id cant found..... Please Enter Corect Product id")
    
    Ui_Quant=int(input("Please Enter Total Quantity Of Prodect"))
    
    check=0
    if(int(recods[Ui_code]["Quantity_in_stock"])==0):
        print("Sorry We are running out of stock for this item")
        check=1
    elif(Ui_Quant>int(recods[Ui_code]["Quantity_in_stock"])):
        print("Sorry we don't have that much of quantity")
        print("WE have only ",recods[Ui_code]["Quantity_in_stock"],"products")
        u=input("Do yoy want to continue with stock avilable at us..... If yes please enter Y")
        if(u=="y" or u=="Y"):
            Ui_Quant=recods[Ui_code]["Quantity_in_stock"]
        else:
            break
    if(check==0):
        prod_amt=Ui_Quant*int(recods[Ui_code]["Unit_Price"])
        amount+=prod_amt;
        convert=str(prod_amt)
        u_list.append("Product : "+recods[Ui_code]['Name']+" , "+"Unit Price - "+recods[Ui_code]["Unit_Price"]+" , "+"Quantity - "+str(Ui_Quant)+" , "+"Amount - "+convert)
    if(int(recods[Ui_code]["Quantity_in_stock"])>0):
        recods[Ui_code]["Quantity_in_stock"]=int(recods[Ui_code]["Quantity_in_stock"])-Ui_Quant
    Ui_c=input("Do you want to order more ?(then please press 1(One) else anykey")
    if(int(Ui_c)==1):
        continue
    elif(int(Ui_c)==0):
        break
    else:
        print("Wrong Input ")
        print("Prociding towards Bill")
        break
import time

dis=0
if(amount<=300):
    dis=0
elif(amount<=500 and amount>=301):
    dis=(3/100)*amount
elif(amount<=2000 and amount>=501):
    dis=(10/100)*amount
else:
    dis=(15/100)*amount
Bill_Num=int(list(stock.keys())[-1])+1
print()
print()
print("********************************************************************")
print("                                Bill                                ")
print("Bill Number -" ,Bill_Num)
print(time.ctime())
print("Customer Name -", u_name,"                 Mob-",U_mob)
for i in u_list: 
    print(i)

print()
print("--------------------------------------------------------------------")
print("                       Total Amount -",amount)
print("                       Discount-","%.2f" % dis)
print("                       Total Payable Amount -",amount-dis)
print("--------------------------------------------------------------------")
print("                               Thank You                            ")
print("                           Please Visit Again                       ")
print("********************************************************************")



js = json.dumps(recods)
fd = open("record.json",'w')
fd.write(js)
fd.close()

stock[Bill_Num] ={"Name":u_name,"Mob":U_mob,"Items":u_list,"Amount":amount-dis,"discount":dis, "date":time.ctime() }

jss = json.dumps(stock)
fd = open("sales.json",'w')
fd.write(jss)
fd.close()
```

```python
print("Here is Your Stocks")
for stock in recods: 
    print("Product -" ,recods[stock]['Name'],",","Product id :",stock,"," , "Unit Price - ₹",recods[stock]["Unit_Price"],",","Item Unit-",recods[stock]["Unit"],",","Quantity in stock",recods[stock]["Quantity_in_stock"],",","Expiry Date -",recods[stock]["Exp_date"],",","Category -",recods[stock]["Category"],",","Description",recods[stock]["Desc"] )
print("Want to add product in Stock or edit already exist product")
print("For Edit press 1")
print("For Add press 0")
a_ip=int(input("Enter your choice "))
if(a_ip==1):
    a_proid= input("Please Enter Product id")
    print(recods[a_proid])
    print("What do you want to update")
    print(" Press 1 to Update Name \n Press 2 to Update Quantity_in_stock \n Press 3 to Update Unit \n Press 4 to Update Unit_Price\n Press 5 to Update Exp_date\n Press 6 to Update Category\n Press 7 to Update Description")
    a_val=int(input("Enter choice"))
    if(a_val==1):
        a_inp1=input("Enter Modified Name")
        recods[a_proid]["Name"]=a_inp1
        print("Updated Sucessfully")
    
    if(a_val==2):
        a_inp2=input("Enter Modified Quantity_in_stock")
        recods[a_proid]["Quantity_in_stock"]=int(a_inp2)+int(recods[a_proid]["Quantity_in_stock"])
        print("Updated Sucessfully")
    
    if(a_val==3):
        a_inp3=input("Enter Modified Unit")
        recods[a_proid]["Unit"]=a_inp3
        print("Updated Sucessfully")
    
    if(a_val==4):
        a_inp4=input("Enter Modified Unit_Price")
        recods[a_proid]["Unit_Price"]=a_inp4
        print("Updated Sucessfully")
    
    if(a_val==5):
        a_inp5=input("Enter Modified Exp_date")
        recods[a_proid]["Exp_date"]=a_inp5
        print("Updated Sucessfully")
        
    if(a_val==6):
        a_inp6=input("Enter Modified Category")
        recods[a_proid]["Category"]=a_inp6
        print("Updated Sucessfully")
        
    if(a_val==7):
        a_inp7=input("Enter Modified Description")
        recods[a_proid]["Desc"]=a_inp7        
        print("Updated Sucessfully")

if(a_ip==0):
    print("Product id Will be auto generated in sequence")
    a_pn=input("Enter Product name")
    a_qun=input("Enter Quantity of product")
    a_un=input("Enter Unit(kg/Piece)")
    a_up=input("Enter Unit Pirce")
    a_exp=input("Enter Expiry Date")
    a_cat=input("Enter category")
    a_desc=input("Enter Description")
    
    last_Product_id=int(list(recods.keys())[-1])+1
    recods[last_Product_id] ={'Name': a_pn, 'Quantity_in_stock': a_qun, 'Unit': a_un, 'Unit_Price': a_up, 'Exp_date': a_exp, 'Category': a_cat, 'Desc': a_desc}
    print("Product Added Sucessfully")
    
js = json.dumps(recods)
fd = open("record.json",'w')
fd.write(js)
fd.close()

```
